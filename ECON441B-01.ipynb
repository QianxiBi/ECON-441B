```python
x = df['alcohol']
y = df['quality']
plt.scatter(x, y)
plt.title('relationshp between alcohol and wine quality')
plt.xlabel('alcohol')
plt.ylabel('wine quality')
plt.show()
```

As we can see from the plot, there tend to be a slight positive relationship between alcohol and wine quality. Wine with higher alcohol tend to be with higher quality. However, the trend is not that obvious and strong.
